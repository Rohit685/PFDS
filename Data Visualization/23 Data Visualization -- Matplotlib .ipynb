#### &emsp;&emsp;Example 1: Basic Histogram

```python
# Use NumPy to create some data

norm_data = np.random.normal(size = 1500, loc = 100, scale = 25)

# plot a histogram using default values

plt.hist(x = norm_data);
```

#### &emsp;&emsp;Example 2: Change the Color of the Bins

```python
# run the same code as in Example 1, but add the color parameter

norm_data = np.random.normal(size = 1500, loc = 100, scale = 25)
plt.hist(x = norm_data, color = 'red');
```

#### &emsp;&emsp;Example 3: Change the Number of Bins

```python
# run the same code as in Example 1, but add the bins parameter

norm_data = np.random.normal(size = 1500, loc = 100, scale = 25)
plt.hist(x = norm_data, color = 'green', bins = 50);
```

### 1.&emsp;PURPOSE OF A PIE CHART

<br>
 <font size="4"><center><i>Pie charts are used to express a part-to-whole relationship. </i></center></font>
<br>

***

<center><b><i>There are two primary use cases for a pie chart:</i></b></center>

1. If you want your audience to have a general sense of the part-to-whole relationship in your data and comparing the precise sizes of the slices is less important.

2. To convey that one segment of the total is relatively small or large.

<a class="anchor" id="52"></a>
### 2.&emsp;SYNTAX OF A MATPLOTLIB PIE CHART

<a class="anchor" id="53"></a>
### 3.&emsp;PARAMETERS OF A MATPLOTLIB PIE CHART

<a class="anchor" id="54"></a>
### 4.&emsp;EXAMPLES OF MATPLOTLIB PIE CHARTS

#### &emsp;&emsp;Example 1: Basic Pie Chart

```python
x = [12, 20, 5, 23, 40]

plt.pie(x);        # By default the plotting of the first wedge starts from the x-axis and moves counterclockwise
                   # The size of each wedge is determined by the formula x/sum(x)
```

#### &emsp;&emsp;Example 2: Add Labels to the Pie Chart

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']

plt.pie(x, labels = our_labels); 
```

#### &emsp;&emsp;Example 3: Make a Wedge Stand Out

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']
our_explode = [0, 0, 0, 0, 0.2]

plt.pie(x, labels = our_labels, explode = our_explode); 
```

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']
our_explode = [0, 0, 0.2, 0, 0.2]

plt.pie(x, labels = our_labels, explode = our_explode); 
```

#### &emsp;&emsp;Example 4: Add a Shadow

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']
our_explode = [0, 0, 0, 0, 0.2]

plt.pie(x, labels = our_labels, explode = our_explode, shadow = True); 
```

#### &emsp;&emsp;Example 5: Set the Colors for Each Wedge

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']
our_colors = ["brown", "g", "pink", "#00ffff", "#458b74" ]

plt.pie(x, labels = our_labels, colors = our_colors); 
```

- You can use [**HTML Color Names**](https://www.w3schools.com/colors/colors_names.asp), [**Hexadecimal Color Values**](https://www.w3schools.com/colors/colors_hexadecimal.asp), or one of the shortcuts below:<br>



#### &emsp;&emsp;Example 6: Add Percents

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']

plt.pie(x, labels = our_labels, autopct = '%1.1f%%');
```

#### &emsp;&emsp;Example 7: Add a Legend

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']

plt.pie(x, labels = our_labels)
plt.legend(); 
```

```python
x = [12, 20, 5, 23, 40]
our_labels = ['Apples', 'Bananas', 'Cherries', 'Grapes', 'Peaches']

plt.pie(x, labels = our_labels)
plt.legend(title = "Fruits"); 
```

```python
# full example

import numpy as np
import matplotlib.pyplot as plt

normal_data = np.random.normal(size = 2000, loc = 50, scale = 15)

plt.style.use('fivethirtyeight')
plt.hist(normal_data);
```
