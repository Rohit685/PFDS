<div class="alert alert-block alert-success"><b>Question 2:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>'Income'</b> of at least 100000. 

```python
customers.loc[customers['Income']>=1000,:]
```

<div class="alert alert-block alert-success"><b>Question 3:</b><br>
    
Select the <b>'Education'</b> Series from the Question 2 filtered DataFrame using the .loc label-based indexing property.
<p>&nbsp;</p>   
<b>Note:</b> See the image below for information on selecting data via the Pandas .loc (label-based indexing) and .iloc (position-based indexing) indexing properties:


    
</div>
    


Documentation for [**'loc'**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html)<br>

<div class="alert alert-block alert-success"><b>Question 4:</b><br>
    
 Select the <b>'Marital_Status'</b> Series from the Question 2 filtered DataFrame without using the .loc label-based indexing attribute.<br>

```python
customers['Marital_Status']
```

<div class="alert alert-block alert-success"><b>Question 5:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>'Income'</b> of at least 80000 and a <b>'Marital_Status'</b> of Single.

```python
customers.loc[(customers['Income']>=80000) & (customers['Marital_Status']=='Single'),:]
```

<div class="alert alert-block alert-success"><b>Question 6:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>'Income'</b> greater than 80000, a <b>'Marital_Status'</b> of Divorced, and an '<b>'Education'</b> level of Graduation.

```python
customers.loc[(customers['Income']>=80000) & (customers['Marital_Status']=='Divorced') & (customers['Education']=='Graduation'),:]
```

<div class="alert alert-block alert-success"><b>Question 7:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with a <b>'Marital_Status'</b> of Married or Together or an '<b>'Education'</b> level of PhD.

```python
customers.loc[(customers['Education']=='PhD') | (customers['Marital_Status']=='Married') | (customers['Marital_Status']=='Together'),:]
```

<div class="alert alert-block alert-success"><b>Question 8:</b><br>
    
Filter the <b>customers</b> DataFrame rows to show only customers with an <b>Income</b> between 40000 and 60000 or an '<b>'Education'</b> level of Master or PhD.

```python
customers.loc[((customers['Income']>=40000) & (customers['Income']<=60000)) | ((customers['Education']=='Master') |(customers['Education']=='PHD')) ,:]
```

<div class="alert alert-block alert-success"><b>Question 9:</b><br>
    
Use the <b>"isin() method"</b> to filter the <b>customers</b> DataFrame to show only customers with a <b>Marital_Status</b> of Single, Married, or Together.<br>
<p>&nbsp;</p>
 
Documentation for [**'isin'**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isin.html)

```python
specific_statuses = customers['Marital_Status'].isin(['Single','Married','Together'])
customers[specific_statuses]
```
