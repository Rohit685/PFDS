<div class="alert alert-block alert-success"><b>Question 9:</b><br>
    
a. Read in the dataset <b>'food_prices.txt'</b> and store the results in a DataFrame named <b>prices</b>.<br> 

See the links below for access to the dataset.

[food_prices.txt](https://drive.google.com/file/d/1nOtnYLxI-kBPAPNKpW9QjCccsOTtTPRl/view?usp=sharing)<br>

<p>&nbsp;</p>
b. Use appropriate methods and attributes to inspect the <b>customers</b> DataFrame. Consider the following options.
 
- head()
- tail()
- info()
- index
- columns
- shape
- dtypes

```python
prices = pd.read_table('food_prices.txt', sep=',')
print(prices.head())
print(prices.tail())
print(prices.info())
print(prices.index)
print(prices.columns)
print(prices.shape)
print(prices.dtypes)
```

<div class="alert alert-block alert-success"><b>Question 10:</b><br>
    
Convert the <b>Price</b> Series of the <b>prices</b> DataFrame from a string to a float.   

```python
prices['Price'] = prices['Price'].apply(lambda x: float(x.split()[0].replace('$', '')))
prices
```

<div class="alert alert-block alert-success"><b>Question 11:</b><br>
Create a new column named <b>'Revenue'</b>. The <b>'Revenue'</b> column can be derived by using the following formula:<br>

Revenue = Price * Quantity Sold

```python
prices['Revenue'] = prices['Price'] * prices['Quantity Sold']
prices
```

<div class="alert alert-block alert-success"><b>Question 12</b><br>
    
Determine the total amount of revenue that was generated by the food sales found in the <b>prices</b> DataFrame.<br>

```python
prices.Revenue.sum().astype(int)
```

<div class="alert alert-block alert-success"><b>Question 13</b><br>

Determine the average price of an item in the <b>prices</b> DataFrame.

```python
print(f"Mean price: {prices.Price.mean():.2f}")
```
