<div class="alert alert-block alert-danger">
<b>Question 16:</b>

Which airline carrier is the fastest? (**Use the distance and airtime variables.**)
    
</div>

```python
flights['DistancePerMin'] = flights.distance / flights.air_time
flights.sort_values('DistancePerMin').max().carrier
```
