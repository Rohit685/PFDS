<div class="alert alert-block alert-success"><b>Question #10:</b> 
The formula for computing the final amount if one is earning compound interest is given on Wikipedia as


<p>&nbsp;</p>   
a. Write a Python program that assigns the principal amount of 10000 to variable P, assigns to n the value 12, and assigns to r the interest rate of 8% (0.08). Then have the program prompt the user for the number of years, t, that the money will be compounded for. 

b. Calculate and print the final amount if the user answers that the number of years is t = 20.

  
</div>


```python
P = 10000
n = 12
r = 0.08
t = int(input("Enter the number of years: "))

equation = P * (1 + (r/n))**(n*t)
print(f"The final amount of money after {t} years is: {equation}")
```
