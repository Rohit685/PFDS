<div class="alert alert-block alert-info">
<b><font size="4">Topic #1: What Are Operators?</font></b></div>

- An operator is a symbol that will perform <b>mathematical</b> operations on <b>variables</b> or on <b>values</b>.

- Operators operate on <b>operands</b> (values) and return a <b>result</b>.

<b>Python has 7 types of operators that you can use:</b>

- Arithmetic Operators
- Relational Operators
- Assignment Operators
- Logical Operators
- Membership Operators
- Identity Operators
- Bitwise Operators

<div class="alert alert-block alert-info">
<b><font size="4">Topic #2: Arithmetic Operators</font></b></div>

```python
# Python Arithmetic Operators ... involving integers

4 + 5
10 - 2
6 * 3
20 / 4

# Python Arithmetic Operators ... using variables

a = 20
b = 5
a + b
a - b
a * b
a / b
```

- Operations involving two integers always produce an integer, except for division (/) which results in a float.

```python
# Python Arithmetic Operators ... involving floats

5 + 3.0
10.0 - 7
2.4 * 5
4.2 / 2.1

# Python Arithmetic Operators ... using variables 

x = 12.5 
y = 5
x + y
x - y
x * y
x / y
```

- Operations involving a float always produce a float.

```python
# Python Arithmetic Operators ... Floor Division / Integer Division (//)

16 // 3

# Use variables

pi = 3.14
e = 2.71
pi // e
```

- If the // division operator is used, the result is an integer value, with truncation (of either int or float type).

```python
# Python Arithmetic Operators ... Modulus (Remainder) Operator (% -- percent sign)

17 % 5

# Use variables

x = 24
y = 7
x % 7
```

- The modulus operator (%) divides one value by another value and gives their remainder.

```python
# Python Arithmetic Operators ... Exponentiation (**)

5 ** 3

# Use variables

a = 3
b = 5
a ** b

# Key point ... successive exponentiation works right to left

2 ** 3 ** 2

# 2 ** 3 ** 2 = 2 ** (3 ** 2)
```

<b>Exponentiation</b> 	

- The right-most ** operator gets done first!
- use parentheses to force the order you want!

<div class="alert alert-block alert-info">
<b><font size="4">Topic #3: RELATIONAL / COMPARISON OPERATORS</font></b></div>

<b>Python has 6 relational operators:</b>

- \> (Greater than)
- < (Less than)
- == (Equal to)
- != (Not equal to)
- \>= (Greater than or equal to)
- <= (Less than or equal to)

```python
# Relational / Comparison Operators

11 > 7
11 < 7
12 >= 5
12 <= 5

# Use some variables
a = 8 + 2
b = 17
a == 10
a != b
```

- Relational operators compare two values and then return either a True or False value.

<div class="alert alert-block alert-info">
<b><font size="4">Topic #4: ASSIGNMENT OPERATORS</font></b></div>

- <b>Assignment operators</b> perform an operation and assign a value.

```python
# The assign operator (=) assigns the value (object) on the right to the variable on the left.

var_1 = 7
print(var_1)

# Add and assign
# The add and assign operator (+=) adds the two values and then assigns the result to the variable on the left.

print()

var_1 = 10
print(var_1)
var_2 = 20
var_1 += var_2
print(var_1)

# Subtract and assign
# The subtract and assign operator (-=) subtracts the value on the right side from the value on the left side, then
# assigns the result to the variable on the left.

print()

var_3 = 45
print(var_3)
var_4 = 25
var_3 -= var_4
print(var_3)

# Multiply and assign
# The multiply and assign operator (*=) aassigns the product to the variable on the left.

print()

var_5 = 6
print(var_5)
var_6 = 10
var_5 *= var_6
print(var_5)
```

```python
# Divide and assign
# The divide and assign operator (/=) aassigns the quotient of two variables to the variable on the left.

x = 100
y = 25
print(x)
x /= y
print(x)

print()

# Modulus and assign
# The modulus and assign operator (%=) performs the modulus calculation on the two values and then assigns the 
# result to the variable on the left

x = 43
y = 10
print(x)
x %= y
print(x)

print()

# Exponentiation and assign 
# The exponentiation and assignment operator (**=) performs exponentiation on the two values and then 
# assigns the result to the first (left) variable.

a = 4
b = 3
print(a)
a **= b
print(a)

print()

# Floor / Integer divide and assign 
# The floor divide and assign operator (//=) performs floor division on the two values and then assigns
# the result to the variable on the left.

m = 57
n = 13
print(m)
m //= n
print(m)
```

<div class="alert alert-block alert-info">
<b><font size="4">Topic #5: LOGICAL OPERATORS</font></b></div>

<b>Python Logical Operators</b> can <b>combine conditions</b>. 

Python has 3 logical operators:

- and (Logical and)
- or (Logical or)
- not (Logical not)

```python
# Logical and
# The Logical operator returns True if both values are True. Otherwise, it returns False.

True and False
True and True
2 < 8 and 5 != 8   
5 and 7       # In situations like this the logical operator and doesn't give True or False -- it gives
              # the last value.
```

```python
# Logical or
# The logical or operator returns True if at least on value is True.  It returns False if both values are False.

True or False
False or True
True or True
False or False
2 > 10 or 7 + 3 == 10
```

```python
# Logical not
# The logical not operator returns True if an expression is False and returns False if an expression is true

not True
not False

x = True
not x

y = False
not y

not 8    # In this situation, 8 is True, so not 8 is False.
```

<div class="alert alert-block alert-info">
<b><font size="4">Topic #6: MEMBERSHIP OPERATORS</font></b></div>

Membership operators check whether a value is in <b>another set of values</b>. 

Python has 2 membership opevalues
    
- in
- not in

```python
# in operator
# The in operator returns True if the first value is in the second set of values. Otherwise, it returns False.

2 in [5, 7, 10]
10 in [5, 10, 15, 20]

"act" in "Facts don't lie."
"snow" in "summer"
```

```python
# not in operator
# The not in operator returns true if the first value is not in the second set of values.  Otherwise it returns False.

2 not in [5, 7, 10]
10 not in [5, 10, 15, 20]

"act" not in "Facts don't lie."
"snow" not in "summer"
```

<div class="alert alert-block alert-info">
<b><font size="4">Topic #7: IDENTITY OPERATORS</font></b></div>

<b>Identity Operators</b>

Identity operators check whether two values are <b>identical</b>. 

Python has 2 identity operators as well:

- is
- is not

```python
# is operator
# The is operator returns True if the first value is the same as the second value.  Otherwise, it returns False.
# ... i.e., it returns True if the two objects have tha same identity.

a = 5
b = a
a is b

x = 7
y = 4 + 5
x is y
```

```python
# is not operator
# The is not operator returns True if the first value is not identical to the second value.  Otherwise, it returns False.
```

```python
a = 5
b = a
a is not b

x = 7
y = 4 + 5
x is not y
```

<b><font size="4">Topic #7: BITWISE OPERATORS</font></b></div>

Bitwise operators operate on values <b>bit by bit</b>.

Python has 6 bitwise operators:

- & (Bitwise and)
- | (Bitwise or)
- ^ (Bitwise xor)
- ~ (Bitwise 1s complement)
- << (Bitwise left-shift)
- \>> (Bitwise right-shift)

At this time no further discussion is needed for Bitwise operators.
