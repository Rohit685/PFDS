<b><font size="4">47. String Methods -- Additional Notes and Examples</font></b></div>

```python
sde = "School District of Elmbrook"
print(sde.upper())

sde_2 = sde.lower()
print(sde_2)

print(sde)
```

- In this example, **upper** is a method that can be invoked on any string object to create a new string in which all the characters are in uppercase. 
<p>&nbsp;</p>
- The method **lower** works in a similar fashion changing all characters in the string to lowercase. (The original string sde remains unchanged. A new string sde_2 is created.)

In addition to **upper** and **lower**, the following table provides a summary of some other useful string methods.
<p>&nbsp;</p>



<p>&nbsp;</p>

You should experiment with these methods so that you understand what they do. Note once again that the methods that return strings do not change the original. You can also consult the [String Methods -- Python documentation](https://docs.python.org/3/library/stdtypes.html#string-methods) for strings.

```python
# String methods examples

city = "    Brookfield, Wisconsin  "

city_i = city.count("i")    # counts the number of i's
print(city_i)               

print(city.strip())         # returns a string with the leading and trailing whitespace removed
print(city.lstrip())        # returns a string with the leading whitespace removed
print(city.rstrip())        # returns a string with the trailing whitespace removed

print("*%!" + city.strip() + "*%!")
print("*%!" + city.lstrip() + "*%!")
print("*%!" + city.rstrip() + "*%!")
```

```python
# More sting methods examples

dinner = "chicken parmesan"
print(dinner.capitalize())         # returns a string with first character capitalized and other characters lowercase

print(dinner.center(30))           # returns a string centered in a field of 30 spaces
print(dinner.ljust(30))            # returns a string left justified in a field of 30 spaces
print(dinner.rjust(30))            # returns a string right justified in a field of 30 spaces

print(dinner.find("ick"))          # returns the leftmost index where the substring item "ick" is found, or -1 if not found
print(dinner.find("c"))            # returns the leftmost index where the substring item "c" is found, or -1 if not found
print(dinner.find("z"))            # returns the leftmost index where the substring item "z" is found, or -1 if not found

print(dinner.index("p"))           # works like find except causes a runtime error if item is not found
```
